```python
#Checkpoint 3: Featues should have proper data types
#'Transaction_time should be in datetimens format.
df_coffee['transaction_time']=pd.to_datetime(df_coffee['transaction_time'], format='%H:%M:%S')
```

```python
#Checkpoint 4: Handling outliers.
num_cols = ['transaction_qty',
            'unit_price',
            'Revenue',
            'Hour']
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df_coffee[col], color='red')
    plt.title(col)
    plt.show()
```

```python
for col in num_cols:
    Q1 = df_coffee[col].quantile(0.25)
    Q3 = df_coffee[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_coffee = df_coffee[
        (df_coffee[col] >= lower) &
        (df_coffee[col] <= upper)
    ]
```
